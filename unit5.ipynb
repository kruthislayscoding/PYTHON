```python
# AIM: Demonstrate the usage of @classmethod and @staticmethod in Python.

# --------------------------------------------------------
# CLASS METHODS
# --------------------------------------------------------

# Class methods operate on the class.
# PURPOSE: They are used to set or get details/status of the class.
# They cannot access or modify specific instance data.
# They are bound to the class instead of objects.
# 
# 2 IMPORTANT THINGS:
# 1. You must specify with @classmethod decorator.
# 2. They take one default parameter 'cls', used for class-level operations.

class Math:
    pi_value = 3.1415  # Class variable

    @classmethod
    def pi(cls):  # Class method
        return cls.pi_value  # Accessing class-level variable

print(Math.pi())  # Calling class method directly without an instance


# --------------------------------------------------------
# STATIC METHODS
# --------------------------------------------------------

# Static methods do not access class or instance data.
# They are defined using @staticmethod.
# They are bound to the class rather than an object.
# 
# PURPOSE: This method can be called without an object for that class.
# They **cannot** modify the state of an object, as they are **not bound** to it.
# They use the @staticmethod decorator.

class MathOperations:
    @staticmethod
    def add(a, b):  # Static method
        return a + b  # Performs an independent operation

print(MathOperations.add(3, 5))  # Calling static method without an instance

#PGM SHOWING BOTH
class Math:
@classmethod
def pi(cls):
return 3.14
@staticmethod
def add(a, b):
return a + b
print(Math.pi())
print(Math.add(3, 5))
# --------------------------------------------------------
# DIFFERENCE BETWEEN CLASS METHOD AND STATIC METHOD
# --------------------------------------------------------
# | Feature           | Class Method                              | Static Method                          |
# |------------------|----------------------------------|----------------------------------|
# | Definition       | Defined using @classmethod      | Defined using @staticmethod     |
# | Bound To        | Class (not instance)            | Class (not instance)            |
# | Takes Parameter | Takes 'cls' as the first param  | No default parameters required  |
# | Access Class Data | Yes, can access class variables | No, cannot access class variables |
# | Access Instance Data | No, cannot access instance data | No, cannot access instance data |
# | Purpose         | Used for class-level operations | Used for independent utility functions |
# | Modification   | Can modify class variables       | Cannot modify class or instance data |
# | Call Without Instance | Yes                           | Yes                           |

```

```python
# AIM: Demonstrate Encapsulation in Python.

# --------------------------------------------------------
# THEORY:
# --------------------------------------------------------
# Encapsulation is the process of bundling data and methods within a single unit.
# A class is an example of encapsulation because it binds data (attributes) 
# and methods (functions) together.
#
# Example: When we create a class, we are implementing encapsulation.
# The class contains attributes (data) and methods (functions) to operate on them.
# --------------------------------------------------------

# Defining a class to demonstrate encapsulation
class Employee:
    def __init__(self, name, project):  # Constructor to initialize attributes
        self.name = name    # Object attribute for name
        self.project = project  # Object attribute for project

    def work(self):  # Method to print employee details
        print(self.name, "is working on", self.project)  

# Creating an instance (object) of Employee class
emp1 = Employee("Alice", "AI Research")  
emp1.work()  # Calling the method work()
```

```python
# --------------------------------------------------------
# Discuss the different access modifiers in Python with examples.
# --------------------------------------------------------

# Access modifiers:
# 1. Public: Accessible WITHIN AND OUTSIDE ALL MEMBER VAR OF CLASS ARE DEF PUBLIC  
# GIVE SAME EMPLOYEE ALICE PGM 

class Employee:
    def __init__(self, name, project):
        self.name = name  # Public variable
        self.project = project  # Public variable

    def work(self):
        print(self.name, "is working on", self.project)

# Object creation
emp1 = Employee("Alice", "AI Research")
print(emp1.name)  # ✅ Accessible (Public)
print(emp1.project)  # ✅ Accessible (Public)
emp1.work()  # ✅ Accessible (Public)


# 2. Protected (_): ACCESIBLE WITHIN CLASS AND ALSO AVAILABLE TO SUBCLASS, 
# TO DEFINE PROTECTED MEMEBER PREFIX MEMBER NAME + SINGLE UNDERSCORE 
# THESE ARE USED WHEN YOU IMPLEMENT INHERITANCE AND WANT TO ALLOW DATA MEM TO ACCESS TO CHILD CLASSES ONLY 

class Employee:
    def __init__(self, name, project):
        self.name = name  # Public variable
        self._project = project  # Protected variable

    def work(self):
        print(self.name, "is working on", self._project)

# Object creation
emp2 = Employee("Alice", "AI Research")
print(emp2.name)  # ✅ Public variable is accessible
print(emp2._project)  # ⚠️ Not recommended, but accessible (Protected)
emp2.work()  # ✅ Accessible


# 3. Private (__): ADD 2 UNDERSCORE AS A PREFIX AT START OF VAR NAME 
# AND ACCESSIBLE ONLY WITHIN CLASS and CANT ACCESS THEM DIRECTLY FROM CLASS OBJESCTS 

class Employee:
    def __init__(self, name, salary):
        self.name = name  # Public variable
        self.__salary = salary  # Private variable

    def show_salary(self):
        print("Salary:", self.__salary)  # ✅ Accessible inside class

# Object creation
emp3 = Employee("Alice", 75000)
print(emp3.name)  # ✅ Public variable is accessible
# print(emp3.__salary)  # ❌ Error: Cannot access private variable directly
print(emp3._Employee__salary)  # ✅ Access using name mangling (Not recommended)
emp3.show_salary()  # ✅ Accessible through method inside class


# THEN FINALLY THIS EXAMPLE GIVEN DOWN GIVE COMMENTS ON TO WHERE WHAT ACCESS MODIFIER HAS BEEN USED
# Example:
class Employee:
    def __init__(self):
        self.__salary = 50000  # Private member

# Object creation
emp = Employee()
# print(emp.__salary)  # ❌ Will cause an error
print(emp._Employee__salary)  # ✅ Access using name mangling
```

```python
# --------------------------------------------------------
# INHERITANCE ALLOWS US TO DEFINE A CLASS THAT INHERITS ALL METHODS AND PROP FROM ANOTHER CLASS  
# PARENT CLASS = CLASS BEING INHERITED FROM, ALSO CALLED BASE CLASS  
# CHILD CLASS = CLASS THAT INHERITS FROM ANOTHER CLASS OR DERIVED CLASS  
# --------------------------------------------------------

# --------------------------------------------------------
# SYNTAX:  
# class BaseClass:  
#     {BODY}  
# class DerivedClass(BaseClass):  
#     {BODY}  
# --------------------------------------------------------

# --------------------------------------------------------
# EXAMPLE:  
# CLASS PERSON WITH FIRST NAME AND LAST NAME  
# STUDENT IS DERIVED CLASS THAT INHERITS FROM PERSON  
# --------------------------------------------------------

# Parent class (Base class)
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname  # Public attribute
        self.lastname = lname   # Public attribute

    def print_name(self):
        print("Full Name:", self.firstname, self.lastname)

# Child class (Derived class) inheriting from Person
class Student(Person):
    def __init__(self, fname, lname, student_id):
        super().__init__(fname, lname)  # Calling parent class constructor
        self.student_id = student_id  # New attribute in child class

    def show_details(self):
        print("Student ID:", self.student_id)

# Object creation
s1 = Student("Alice", "Johnson", "S1234")

# Access parent class method
s1.print_name()  # ✅ Inherited method from Person

# Access child class method
s1.show_details()  # ✅ Method defined in Student class
# --------------------------------------------------------
# USING INHERITANCE IN PYTHON TO CREATE NEW CLASSES  
# DERIVED CLASS EXTENDS PARENT CLASS THROUGH INHERITANCE  
# --------------------------------------------------------

# Parent class (Base class)
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname  # Public attribute
        self.lastname = lname   # Public attribute

    def print_name(self):
        print("Full Name:", self.firstname, self.lastname)

# Child class (Derived class) inheriting from Person
class Student(Person):
    def __init__(self, fname, lname, student_id):
        super().__init__(fname, lname)  # Calling parent class constructor
        self.student_id = student_id  # New attribute in child class

    def show_details(self):
        print("Student ID:", self.student_id)

# Further extending Student class to create another derived class
class GraduateStudent(Student):
    def __init__(self, fname, lname, student_id, degree):
        super().__init__(fname, lname, student_id)  # Calling Student class constructor
        self.degree = degree  # New attribute in GraduateStudent

    def show_degree(self):
        print(self.firstname, "is pursuing", self.degree)

# Object creation
g1 = GraduateStudent("Alice", "Johnson", "S1234", "M.Sc. Computer Science")

# Access methods from all parent classes
g1.print_name()      # ✅ Inherited from Person
g1.show_details()    # ✅ Inherited from Student
g1.show_degree()     # ✅ Defined in GraduateStudent
```

```python
# --------------------------------------
# POLYMORPHISM IN PYTHON
# --------------------------------------
# Polymorphism refers to the use of a single type entity (method/operator/object) 
# to represent different types in different scenarios.

# EXAMPLE 1: OPERATOR OVERLOADING
# --------------------------------------
# The '+' operator is used for both:
# (1) Addition of numbers
# (2) Concatenation of strings

num_sum = 10 + 20  # Addition of integers
str_concat = "Hello" + " World"  # Concatenation of strings

print("Sum of numbers:", num_sum)  # Output: 30
print("String concatenation:", str_concat)  # Output: Hello World


# --------------------------------------
# FUNCTION POLYMORPHISM
# --------------------------------------
# Some built-in functions in Python are compatible with multiple data types.
# Example: The len() function can be used for strings, lists, and dictionaries.

string_ex = "Python"
list_ex = [10, 20, 30, 40]
dict_ex = {"name": "Alice", "age": 25}

print("Length of string:", len(string_ex))  # len() for string
print("Length of list:", len(list_ex))  # len() for list
print("Length of dictionary:", len(dict_ex))  # len() for dictionary


# --------------------------------------
# CLASS POLYMORPHISM
# --------------------------------------
# Polymorphism while creating class methods in Python allows 
# different classes to have methods with the same name.

class India:
    def info(self):
        print("India: Capital - New Delhi, Language - Hindi, Status - Developing")

class USA:
    def info(self):
        print("USA: Capital - Washington D.C., Language - English, Status - Developed")

# --------------------------------------
# CLASS POLYMORPHISM WHILE ACCESSING OBJECTS
# --------------------------------------
# A single function can handle different object types.

# Creating objects of both classes
country1 = India()
country2 = USA()

# Accessing the method using polymorphism
for country in (country1, country2):
    country.info()  # Calls info() for India and USA dynamically THEY SHARE SIMILAR STRUCTURE+SAME METHOD NAMES INFO HERE NOTICE WE HAVENT CREATED A 
    #COMMON SUPER CLASS OR LINKED THEM TOGETHER IN ANY WAY EVEN THEN WE CAN PACK 2 OBJ IN A TUPLE AND ITERATE USING COMMON COUNTRY VAR

```

```python
# --------------------------------------
# POLYMORPHISM & INHERITANCE IN PYTHON
# --------------------------------------
# Inheritance: Allows a child class to inherit properties from a parent class.
# Polymorphism: Methods with the same name behave differently in different classes.

# Base class (Parent class)
class Bird:
    def __init__(self, name):
        self.name = name
    
    def fly(self):
        return "Some birds can fly, some cannot."

    def make_sound(self):
        return "Birds make different sounds."

# Derived class 1 (Inherits from Bird)
class Sparrow(Bird):
    def fly(self):  # Overriding the fly method
        return f"{self.name} can fly high!"

    def make_sound(self):  # Overriding make_sound method
        return f"{self.name} chirps!"

# Derived class 2 (Inherits from Bird)
class Ostrich(Bird):
    def fly(self):  # Overriding the fly method
        return f"{self.name} cannot fly but can run fast!"

    def make_sound(self):  # Overriding make_sound method
        return f"{self.name} makes booming sounds!"

# Polymorphism: Different classes using the same method names
birds = [Sparrow("Sparrow"), Ostrich("Ostrich")]

for bird in birds:
    print(bird.fly())  # Calls fly() based on the object type
    print(bird.make_sound())  # Calls make_sound() based on the object type
```

```python
# ------------------------------------------------------------------
# MEMBERSHIP OPERATORS, IDENTITY OPERATORS, AND BITWISE OPERATORS
# ------------------------------------------------------------------

# ------------------------------
# MEMBERSHIP OPERATORS (in, not in)
# ------------------------------
# Membership operators check whether a value exists in a sequence (list, tuple, string, dictionary).
# 'in' returns True if the value is present.
# 'not in' returns True if the value is not present.

my_list = [10, 20, 30, 40, 50]
print(30 in my_list)  # True (30 is present in the list)
print(100 not in my_list)  # True (100 is not in the list)

my_string = "Python"
print("y" in my_string)  # True ('y' is present in "Python")
print("z" not in my_string)  # True ('z' is not in "Python")

# ------------------------------
# IDENTITY OPERATORS (is, is not)
# ------------------------------
# Identity operators check whether two objects refer to the same memory location.
# 'is' returns True if both variables point to the same object.
# 'is not' returns True if they do not point to the same object.

a = [1, 2, 3]
b = a  # Both 'a' and 'b' refer to the same list object
c = [1, 2, 3]  # A new list with the same values but a different object

print(a is b)  # True (both refer to the same object)
print(a is c)  # False (c is a different object)
print(a is not c)  # True (a and c are different objects)

x = 10
y = 10
print(x is y)  # True (small integers are cached, so x and y refer to the same memory)
z = 1000
w = 1000
print(z is w)  # False (large integers are stored separately)

# ------------------------------
# BITWISE OPERATORS (&, |, ^, ~, <<, >>)
# ------------------------------
# Bitwise operators operate on binary representations of numbers.

# Bitwise AND (&): Sets bits to 1 where both operands have 1
print(5 & 3)  # 5 = 0101, 3 = 0011 -> (0101 & 0011) = 0001 (1)

# Bitwise OR (|): Sets bits to 1 if at least one operand has 1
print(5 | 3)  # 5 = 0101, 3 = 0011 -> (0101 | 0011) = 0111 (7)

# Bitwise XOR (^): Sets bits to 1 where only one operand has 1
print(5 ^ 3)  # 5 = 0101, 3 = 0011 -> (0101 ^ 0011) = 0110 (6)

# Bitwise NOT (~): Inverts all bits (two's complement: -(x+1))
print(~5)  # 5 = 0101 -> ~5 = -(5+1) = -6

# Left Shift (<<): Shifts bits left, multiplying by 2
print(5 << 1)  # 5 = 0101 -> (0101 << 1) = 1010 (10)

# Right Shift (>>): Shifts bits right, dividing by 2
print(5 >> 1)  # 5 = 0101 -> (0101 >> 1) = 0010 (2)
```
