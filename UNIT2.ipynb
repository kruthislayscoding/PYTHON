```python
# Python program to illustrate if conditions VERY IMP

# Simple if statement
x = 10
if x > 5:  # If x is greater than 5, this block executes
    print("x is greater than 5")

# If-else statement with syntax and flow
# Syntax:
# if condition:  # If condition is true, execute this block
#     statement
# else:  # If condition is false, execute this block
#     statement

x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")

# If-elif-else statement with syntax and flow
# Syntax:
# if condition1:  # If condition1 is true, execute this block
#     statement1
# elif condition2:  # If condition1 is false but condition2 is true, execute this block
#     statement2
# else:  # If both conditions are false, execute this block
#     statement3

x = 7
if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5 but not greater than 10")
else:
    print("x is 5 or less")

# Elif Ladder example
# Syntax:
# if condition1:  # If condition1 is true, execute this block
#     statement1
# elif condition2:  # If condition1 is false but condition2 is true, execute this block
#     statement2
# elif condition3:  # If all previous conditions are false but condition3 is true, execute this block
#     statement3
# else:  # If none of the above conditions are true, execute this block
#     statement4

x = 50
if x < 20:
    print("x is less than 20")
elif x < 40:
    print("x is between 20 and 39")
elif x < 60:
    print("x is between 40 and 59")
else:
    print("x is 60 or more")

# Nested if-else statement with syntax and flow
# Syntax:
# if condition1:  # If condition1 is true, go inside this block
#     if condition2:  # If condition2 is also true, execute this block
#         statement
#     else:  # If condition2 is false, execute this block
#         statement
# else:  # If condition1 is false, execute this block
#     statement

x = 15
if x > 10:
    if x < 20:
        print("x is between 10 and 20")
    else:
        print("x is greater than or equal to 20")
else:
    print("x is 10 or less")
```

```python
# Program to Check Password Validity and Matching
# AIM: This program validates a user-entered password based on security criteria.
# The second part checks if the re-entered password matches the original.

# Step 1: Check password validity based on length, case, digits, and special characters
password = input("Enter password: ")

# Conditions for a strong password:
# 1. Length must be at least 8 characters
# 2. Must contain at least one uppercase letter
# 3. Must contain at least one lowercase letter
# 4. Must contain at least one digit
# 5. Must contain at least one special character (@, $, !, %, *, ?, &)
if (len(password) >= 8 and 
    any(char.isupper() for char in password) and 
    any(char.islower() for char in password) and 
    any(char.isdigit() for char in password) and 
    any(char in "@$!%*?&" for char in password)):
    print("Valid Password - Access Granted")
else:
    print("Invalid Password - Access Denied")

# Step 2: Check if re-entered password matches the first one
confirm_password = input("Re-enter password: ")

# If passwords match, access is granted; otherwise, denied.
if password == confirm_password:
    print("Passwords Match - Access Granted")
else:
    print("Passwords Do Not Match - Access Denied")
```

```python
# Program to print the sum of all odd numbers from 0 to 50
# Hint: Odd numbers have remainder 1 when divided by 2

sum_odd = 0  # Initialize sum

# Loop through numbers from 0 to 50
for n in range(0, 51):
    if n % 2 != 0:  # Check if the number is odd
        sum_odd += n  # Add odd number to sum

print("Sum of all odd numbers from 0 to 50:", sum_odd)
```

```python
# Theory:
# A while loop executes repeatedly as long as the condition remains True.
# It is useful when the number of iterations is unknown beforehand.

# Flow Diagram:
#         +-------------------+
#         |   Start Loop      |
#         +-------------------+
#                 |
#         +-------------------+
#         |  Test Condition   | 
#         +-------------------+
#            /         \  
#          T           F
#         /             \
# +----------------+    +-----------------+
# |  Execute Body  |    | Exit Loop       |
# +----------------+    +-----------------+
#         |
#  +-------------------+
#  |  Repeat Process  |
#  +-------------------+

# Syntax of while loop:
# while condition:
#     # Code block (executes if condition is True)
#     # Update condition to avoid infinite loop

# ----------------------------------------------------------------------
# Sentry Variable Example : while loops are controlled by a sentry or guard variable, var is used in condn and compared to some other value
# sentry variable=guard if sentry doesnt have value when cond evaluated returns error usually intiated right before loop runs
#make sure its possible for the while to evaluate to true at some point or else it will never run
# The sentry variable 'response' controls whether the loop executes.

print("Welcome!")  # Initial print statement
response = ""  # Sentry variable initialized before loop

while response != "because":  # The loop runs until user enters "because"
    response = input("but why")  # Inside the loop, response is changed to "input of user"
print("Oh okay!")  # Prints when the loop exits

# Explanation:
# - The sentry variable is checked before the loop executes.
# - The loop runs **as long as response is not "because"**.
# - response is used in condn and compared to str"because" before block is executed each time
# - The loop exits only if response is somehow set to `"because"`, printing "Oh okay!".

# ----------------------------------------------------------------------
# ERROR CASE: If the sentry variable is initialized incorrectly, the loop never runs.

response = "because"  # Condition is already False, so loop won't execute
while response != "because":
    response = "but why"
    print("This won't execute!")  # Never runs because condition is False at start

# ----------------------------------------------------------------------
# INFINITE WHILE LOOP (Common mistake)
# This loop runs forever since the condition never becomes False.
# Uncomment the below code to test in a controlled environment.

# counter = 1  # Sentry variable
# while counter > 0:  # Always True because counter is always > 0
#     print("This is an infinite loop!")  # This will keep printing
#     #
```

```python
# Arithmetic Operators
a, b = 10, 3
print("Addition:", a + b)       # 10 + 3 = 13
print("Subtraction:", a - b)    # 10 - 3 = 7
print("Multiplication:", a * b) # 10 * 3 = 30
print("Division:", a / b)       # 10 / 3 = 3.333...
print("Floor Division:", a // b)# 10 // 3 = 3
print("Modulus:", a % b)        # 10 % 3 = 1
print("Exponentiation:", a ** b)# 10^3 = 1000

# Assignment Operators
x = 5
x += 2   # x = x + 2
print("After +=:", x)
x -= 2   # x = x - 2
print("After -=:", x)
x *= 2   # x = x * 2
print("After *=:", x)
x /= 2   # x = x / 2
print("After /=:", x)
x %= 3   # x = x % 3
print("After %=:", x)
x **= 2  # x = x ** 2
print("After **=:", x)
x //= 2  # x = x // 2
print("After //=:", x)

# Comparison Operators
print("Equal:", a == b)         # False
print("Not Equal:", a != b)     # True
print("Greater than:", a > b)   # True
print("Less than:", a < b)      # False
print("Greater or Equal:", a >= b) # True
print("Less or Equal:", a <= b) # False

# Identity Operators
c = [1, 2, 3]
d = c
e = [1, 2, 3]
print("c is d:", c is d)   # True (same object)
print("c is e:", c is e)   # False (different objects)
print("c is not e:", c is not e)  # True

# Membership Operators
print("2 in c:", 2 in c)   # True
print("5 not in c:", 5 not in c)  # True
```
