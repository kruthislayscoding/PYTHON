```python
## Python program to illustrate various escape sequences and string operations

# 1) PRINT FUNCTION: Prints newline character as final value,
# meaning subsequent call to print would print on next line.
# Allows you to specify your own string to be printed without newline.
print('HERE', end='')  # Output: HEREIT IS
print('IT IS')

# 2) TRIPLE QUOTED STRINGS: Used for multi-line string creation and multi-line comments.
multiline_str = """This is a
multi-line string."""
print(multiline_str)  # Output:
# This is a
# multi-line string.

# 3) MOVE FORWARD BY TAB: Good for setting off text and arranging text.
print("Column1\tColumn2\tColumn3")  # Output: Column1    Column2    Column3

# 4) USE 2 BACKSLASHES IN A ROW
print("This is a triple backslash: \\\\")  # Output: This is a triple backslash: \\

# 5) INSERT NEWLINE
print("First line\nSecond line")  # Output:
# First line
# Second line

# 6) INSERT QUOTE
print("This is a quote: '\"'")  # Output: This is a quote: '"

# 7) CONCATENATE STRINGS
str1 = "Hello, "
str2 = "World!"
print(str1 + str2)  # Output: Hello, World!

# 8) LINE CONTINUATION CHARACTER
long_string = "This is a very long string that needs to " \
              "be split across multiple lines."
print(long_string)  # Output: This is a very long string that needs to be split across multiple lines.

# 9) REPEAT STRING
print("Repeat! " * 3)  # Output: Repeat! Repeat! Repeat!
# 10) TYPE CONVERSION
print(float(10), complex(10), int("50"))  # Output: 10.0 (10+0j) 50

# 11) RANDOM NUMBER GENERATION
import random
print(random.randrange(1, 1000))  # Output: Random number between 1 and 1000
```

```python
# Python program to illustrate string methods

s = "Sun is the star"
print("Original String:", s)  # Output: Sun is the star

# upper(): Converts all characters to uppercase.
print(s.upper())  # Output: SUN IS THE STAR

# lower(): Converts all characters to lowercase.
print(s.lower())  # Output: sun is the star

# title(): Returns a string where the first letter of every word is capitalized.
print(s.title())  # Output: Sun Is The Star

# replace(): Replaces occurrences of a substring with another substring.
print(s.replace("the", "a"))  # Output: Sun is a star

# swapcase(): Returns a string with uppercase letters converted to lowercase and vice versa.
print(s.swapcase())  # Output: sUN IS THE STAR

# capitalize(): Returns a string with the first letter capitalized and all others lowercase.
print(s.capitalize())  # Output: Sun is the star

# replace() follows syntax: str.replace(old, new, max)
# It replaces up to 'max' occurrences of 'old' with 'new'.
print(s.replace("is", "was", 1))  # Output: Sun was the star

# strip(): Removes leading and trailing whitespace from the string.
s2 = "  Sun is the star  "
print(s2.strip())  # Output: Sun is the star
```

```python
# Python program to illustrate augmented assignment operators
x = 10
print("Initial x:", x)
x += 5  # Addition (x = x + 5)
print("After x += 5:", x)
x -= 3  # Subtraction (x = x - 3)
print("After x -= 3:", x)
x *= 2  # Multiplication (x = x * 2)
print("After x *= 2:", x)
x /= 4  # Division (x = x / 4)
print("After x /= 4:", x)
x //= 2  # Floor Division (x = x // 2)
print("After x //= 2:", x)
x **= 3  # Exponentiation (x = x ** 3)
print("After x **= 3:", x)
x %= 5  # Modulus (x = x % 5)
print("After x %= 5:", x)
x &= 3  # Bitwise AND (x = x & 3)
print("After x &= 3:", x)
x |= 2  # Bitwise OR (x = x | 2)
print("After x |= 2:", x)
x ^= 1  # Bitwise XOR (x = x ^ 1)
print("After x ^= 1:", x)
x <<= 1  # Left Shift (x = x << 1)
print("After x <<= 1:", x)
x >>= 1  # Right Shift (x = x >> 1)
print("After x >>= 1:", x)
```
